# Convert all ISRCs(from dataset) to Spotify tracki IDs (from spotify API)

##### Connection to the APIs

```python
# --- Spotify credentials
#client_id = "ec41384e2e9140539ea838f37b96fb0e"
#client_secret = "6b4e8c05530a436286ef9f14b12d96a1"

client_id = "YOUR_API_KEY"
client_secret = "YOUR_API_KEY"

auth_string = f"{client_id}:{client_secret}"
auth_base64 = base64.b64encode(auth_string.encode()).decode()

def get_token():
    response = requests.post(
        "https://accounts.spotify.com/api/token",
        headers={"Authorization": f"Basic {auth_base64}"},
        data={"grant_type": "client_credentials"}
    )
    return response.json()["access_token"]



# --- MUSICAE AUTH ---

musicae_headers = {
    "x-rapidapi-key": "YOUR_API_KEY", 
    "x-rapidapi-host": "spotify-extended-audio-features-api.p.rapidapi.com",
    "Content-Type": "application/json"
}

access_token = get_token()
print("Both APIs ready!")
```

```python
# searching for the ISRC
def get_spotify_id(isrc, token):
    """Search one ISRC, return Spotify ID or None"""
    response = requests.get(
        "https://api.spotify.com/v1/search",
        headers={"Authorization": f"Bearer {token}"},
        params={"q": f"isrc:{isrc}", "type": "track", "limit": 1}
    )
    
    if response.status_code == 401:
        return "TOKEN_EXPIRED"
    
    items = response.json()["tracks"]["items"]
    return items[0]["id"] if items else None
```

```python
# getting Spotify API : ISRC -> spotify ID
def get_spotify_id(isrc, token):
    response = requests.get(
        "https://api.spotify.com/v1/search",
        headers={"Authorization": f"Bearer {token}"},
        params={"q": f"isrc:{isrc}", "type": "track", "limit": 1}
    )
    if response.status_code == 401:
        return "TOKEN_EXPIRED"
    items = response.json()["tracks"]["items"]
    return items[0]["id"] if items else None

# Load what we already have
already_done = pd.read_csv("isrc_to_spotify_id.csv")
results = already_done.to_dict("records")
print(f"Starting from row 4101. Already have {len(results)} tracks")

failed_search = []

# START AT 4101 instead of 0
for i in range(4101, len(df)):
    
    isrc = df["ISRC"].iloc[i]  # adjust column name!
    
    # Skip bad ISRCs
    if pd.isna(isrc) or str(isrc).strip() == "":
        failed_search.append(isrc)
        continue
    
    try:
        spotify_id = get_spotify_id(isrc, access_token)
        
        if spotify_id == "TOKEN_EXPIRED":
            access_token = get_token()
            spotify_id = get_spotify_id(isrc, access_token)
        
        if spotify_id:
            results.append({"isrc": isrc, "spotify_id": spotify_id})
        else:
            failed_search.append(isrc)
    
    except Exception as e:
        print(f"\nError at track {i}: {type(e).__name__}: {e}")
        print(f"ISRC was: {isrc}")
        pd.DataFrame(results).to_csv("isrc_to_spotify_id.csv", index=False)
        time.sleep(20)
        access_token = get_token()
        continue
    
    time.sleep(0.5)  # slower to avoid rate limits
    
    if (i + 1) % 300 == 0:
        pd.DataFrame(results).to_csv("isrc_to_spotify_id.csv", index=False)
        print(f"Done: {i+1}/{len(df)} | Matched: {len(results)} | Saved!")

pd.DataFrame(results).to_csv("isrc_to_spotify_id.csv", index=False)
print(f"\nDONE! Matched: {len(results)}/{len(df)} | Failed: {len(failed_search)}")
```

```python
#Rapid API spotify ID -> track features
mapping = pd.read_csv("isrc_to_spotify_id.csv")
all_ids = mapping["spotify_id"].tolist()
print(f"Total IDs to fetch: {len(all_ids)}")

batches = [all_ids[i:i+5] for i in range(0, len(all_ids), 5)] # Batch in groups of 5
print(f"Total batches: {len(batches)}")
```

```python
# Resume from where you left off
START_BATCH = 0  # change this if resuming!

try:
    existing = pd.read_csv("audio_features.csv")
    all_features = existing.to_dict("records")
    # Clean out any None values from previous runs
    all_features = [f for f in all_features if f is not None]
    START_BATCH = len(all_features) // 5
    print(f"Resuming from batch {START_BATCH}. Have {len(all_features)} tracks")
except:
    all_features = []
    print("Starting fresh")

for i in range(START_BATCH, len(batches)):
    try:
        ids_string = ",".join(batches[i])
        
        response = requests.get(
            "https://spotify-extended-audio-features-api.p.rapidapi.com/v1/audio-features",
            headers=musicae_headers,
            params={"ids": ids_string}
        )
        
        data = response.json()
        
        if "audio_features" in data:
            valid = [f for f in data["audio_features"] if f is not None]
            all_features.extend(valid)
            print(f"Batch {i+1}/{len(batches)}: ✓ {len(valid)} tracks | Total: {len(all_features)}")

        else:
            print(f"Batch {i+1}/{len(batches)}: ✗ failed - {str(data)[:100]}")
            pd.DataFrame(all_features).to_csv("audio_features.csv", index=False)
            print(f"Saved {len(all_features)} tracks. Re-run tomorrow!")
            break
    except:
        pd.DataFrame(all_features).to_csv("audio_features.csv", index=False)
        print(f"Error. Saved {len(all_features)} tracks. Re-run tomorrow!")
        break
    
    time.sleep(0.5)  # 2 seconds between requests
    
    if len(all_features) % 100 == 0:
        pd.DataFrame(all_features).to_csv("audio_features.csv", index=False)
        print(f"Progress: {len(all_features)} tracks | Saved!")

pd.DataFrame(all_features).to_csv("audio_features.csv", index=False)
print(f"\nTotal: {len(all_features)} tracks with features")
```
