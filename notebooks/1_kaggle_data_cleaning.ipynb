# Spotify Dataset

```python
import pandas as pd
from pathlib import Path
```

```python
data_path = Path("spotify_dataset.csv")
df = pd.read_csv(data_path, encoding="latin-1")
df.shape
```

```python
cols_to_drop = [
    "Album Name",
    "TIDAL Popularity",
    "TikTok Posts",
    "YouTube Playlist Reach",
    "SiriusXM Spins",
    "Deezer Playlist Count",
    "Deezer Playlist Reach",
    "Amazon Playlist Count",
    "Pandora Streams",
    "Pandora Track Stations",
    "Soundcloud Streams",
]
df = df.drop(columns=cols_to_drop)
df.shape
```

```python
df.columns
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.head()
```

```python
import re


def clean_track_name(name):
    if pd.isna(name):
        return name
    s = str(name)
    s = re.sub(r"\s*\(feat\.?[^)]*\)", "", s, flags=re.IGNORECASE)
    s = re.sub(r"\s*\(ft\.?[^)]*\)", "", s, flags=re.IGNORECASE)
    s = re.sub(r"[^a-zA-Z0-9\s]", "", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s.title()


df["Track"] = df["Track"].apply(clean_track_name)
df["Track"].head(20)
```

```python
numeric_str_cols = [
    "All Time Rank",
    "Spotify Streams",
    "Spotify Playlist Count",
    "Spotify Playlist Reach",
    "YouTube Views",
    "YouTube Likes",
    "TikTok Likes",
    "TikTok Views",
    "AirPlay Spins",
    "Shazam Counts",
]

for col in numeric_str_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", "", regex=False)
        .astype("int64")
    )

int_float_cols = ["Spotify Popularity", "Apple Music Playlist Count", "Explicit Track"]
for col in int_float_cols:
    df[col] = df[col].astype("int64")

df.dtypes
```

```python
df.head()
```

```python
df.to_csv("spotify_kaggle_dataset_clean.csv", index=False)
```
